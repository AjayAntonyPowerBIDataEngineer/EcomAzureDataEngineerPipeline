```python
from pyspark.sql import functions as F
from delta.tables import DeltaTable
from pyspark.sql import functions as F
from pyspark.sql.types import *
```

```python
df = spark.read.table("ecom.silver.dim_silver_products")
df.printSchema()
```

```python
df.select(
    F.col("product_id"),
    F.col("product_category_name"),
    F.col("product_name_lenght"),
    F.col("product_description_lenght"),
    F.col("product_photos_qty"),
    F.col("product_weight_g"),
    F.col("product_length_cm"),
    F.col("product_height_cm"),
    F.col("product_width_cm"),
    F.col("product_volume_cm3"),
    F.col("weight_category"),
)
```

```python
tablename = "ecom.gold.gold_products"
(
        df
        .write
        .format("delta")
        .mode("overwrite")
        .option("overwriteSchema", "true")
        .saveAsTable(tablename)
    )

print("Gold  Load Completed Successfully")
```
