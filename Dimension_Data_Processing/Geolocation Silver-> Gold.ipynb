```python
%sql
SELECT * FROM ecom.silver.dim_silver_geo
```

```python
df = spark.read.table("ecom.silver.dim_silver_geo")
df.printSchema()
```

```python
from pyspark.sql.functions import col
df.select(
    col("geolocation_zip_code_prefix"),
    col("geolocation_lat"),
    col("geolocation_lng"),
    col("geolocation_city"),
    col("geolocation_state")
)
```

```python
tablename = "ecom.gold.gold_geo"
```

```python
(
        df
        .write
        .format("delta")
        .mode("overwrite")
        .option("overwriteSchema", "true")
        .saveAsTable(tablename)
    )

print("Full Load Completed Successfully")
```
