```python
from pyspark.sql import functions as F
from delta.tables import DeltaTable
from pyspark.sql import functions as F
from pyspark.sql.types import *
```

```python
df = spark.read.table("ecom.silver.dim_silver_sellers")
df.printSchema()
```

```python
df.select(
    F.col("seller_id"),
    F.col("seller_dim_id"),
    F.col("seller_zip_code_prefix"),
    F.col("seller_city"),
    F.col("seller_state"),
    F.col("start_date"),
    F.col("end_date") 
)
```

```python
tablename = "ecom.gold.gold_sellers"
(
        df
        .write
        .format("delta")
        .mode("overwrite")
        .option("overwriteSchema", "true")
        .saveAsTable(tablename)
    )

print("Gold  Load Completed Successfully")
```
