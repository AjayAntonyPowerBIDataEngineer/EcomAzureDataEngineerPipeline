```python
from pyspark.sql import functions as F
from delta.tables import DeltaTable
from pyspark.sql import functions as F
from pyspark.sql.types import *
```

```python
%sql
SELECT * FROM ecom.silver.dim_silver_customers
```

```python
df = spark.read.table("ecom.silver.dim_silver_customers")
df.printSchema()
```

```python
from pyspark.sql.functions import col
df.select(
    col("customer_dim_id"),
    col("customer_id"),
    col("customer_unique_id"),
    col("customer_city"),
    col("customer_state"),
    col("customer_zip_code_prefix"),
    col("is_active"),
    col("start_date"),
    col("end_date")
)
```

```python
tablename = "ecom.gold.gold_customers"
```

```python
(
        df
        .write
        .format("delta")
        .mode("overwrite")
        .option("overwriteSchema", "true")
        .saveAsTable(tablename)
    )

print("Full Load Completed Successfully")
```
